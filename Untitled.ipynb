```python
#Q1: Given two arrays, write a python function to return the intersection of the two? For example, X = [1,5,9,0] and Y = [3,0,2,9] it should return [9,0]

def intersection(x,y):
    return list(set(x)&set(y))
x=[1,5,9,0]
y= [3,0,2,9]

z = intersection(x,y)
print(z)
```

```python
#Q2: Given an array, find all the duplicates in this array? For example: input: [1,2,3,1,3,6,5] output: [1,3]

x = [1,2,3,1,3,6,5,1] 
set1 = set()
res_set = set()

for i in x:
    if i in set1:
        res_set.add(i)
    else:
        set1.add(i)
        
print(res_set)
```

```python
#Given an integer array, find the sum of the largest contiguous subarray within the array. For example, given the array A = [0,-1,-5,-2,3,14] it should return 17 because of [3,14]. Note that if all the elements are negative it should return zero.
a = [0,-1,-5,-2,3,14]
def largest_contiguous_array(a):
    max_sum=a[0]
    curr_sum = 0
    for i in range(len(a)):
        if max_sum<i:
            curr_sum += a[i]
            max_sum = max(max_sum,curr_sum)
            if curr_sum<0:
                curr_sum =0
    return max_sum
print(largest_contiguous_array(a))
```

```python
#Q7: Given an integer n and an integer K, output a list of all of the combination of k numbers chosen from 1 to n. For example, if n=3 and k=2, return [1,2],[1,3],[2,3]
from itertools import combinations

def generate_combinations(n, k):
    return list(combinations(range(1, n+1), k))

# Example Usage
n, k = 3, 2
result = generate_combinations(n, k)
print(result)  # Output: [(1, 2), (1, 3), (2, 3)]
```

```python
#Q8: Write a function to generate N samples from a normal distribution and plot them on the histogram
import numpy as np
import matplotlib.pyplot as plt

def plot_normal_distribution(n, mean=0, std_dev=1):
    # Generate N samples from a normal distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=n)

    # Plot histogram
    plt.hist(samples, bins=30, edgecolor='black', alpha=0.7, density=True)
    
    # Add title and labels
    plt.title(f'Normal Distribution (mean={mean}, std_dev={std_dev}, N={n})')
    plt.xlabel('Value')
    plt.ylabel('Density')
    
    # Show plot
    plt.show()
```

```python
#Q10: Given a string, return the first recurring character in it, or “None” if there is no recurring character. Example: input = "pythoninterviewquestion" , output = "n"
input_string = "pythoninterviewquestion"

def first_recurring(input_str):
  
  a_str = ""
  for letter in input_str:
    a_str = a_str + letter
    if a_str.count(letter) > 1:
      return letter
  return None

first_recurring(input_string)
    
```
