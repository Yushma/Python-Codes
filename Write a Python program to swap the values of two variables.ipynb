# Write a Python program to swap the values of two variables.

```python
var1 = input("Enter first variable ")
var2 = input("Enter second variable")

print(f"Values of first and second variables before swaping var1 = {var1} and var2 = {var2}")

var1, var2 = var2, var1

print(f"Values of first and second variables after swaping var1 = {var1} and var2 = {var2}")
 
```
