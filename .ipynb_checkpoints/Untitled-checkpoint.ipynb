```python
#Q1: Given two arrays, write a python function to return the intersection of the two? For example, X = [1,5,9,0] and Y = [3,0,2,9] it should return [9,0]

def intersection(x,y):
    return list(set(x)&set(y))
x=[1,5,9,0]
y= [3,0,2,9]

z = intersection(x,y)
print(z)
```

```python
#Q2: Given an array, find all the duplicates in this array? For example: input: [1,2,3,1,3,6,5] output: [1,3]

x = [1,2,3,1,3,6,5,1] 
set1 = set()
res_set = set()

for i in x:
    if i in set1:
        res_set.add(i)
    else:
        set1.add(i)
        
print(res_set)
```

```python
#Given an integer array, return the maximum product of any three numbers in the array?
def intersection(arr1, arr2):
    return list(set(arr1) & set(arr2))  # Convert lists to sets and find intersection

# Example Usage
X = [1, 5, 9, 0]
Y = [3, 0, 2, 9]
result = intersection(X, Y)
print(result)  # Output: [0, 9] (Order may vary)
```
