```python
#Given a tuple arr , print "True" if all elements of tuple are different otherwise print "False".A tuple is a collection of items that are ordered and unchangeable.
```

```python
def dist_tuple(arr):
    if len(arr) == len(set(arr)):
        print("True")
    else:
        print("False")
        
arr =(1,2,3,5,4,6,3)
dist_tuple(arr)        
```
