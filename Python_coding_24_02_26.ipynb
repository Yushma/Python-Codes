# Reverse string

```python
def reverse_string(s):
    return s[::-1]
print(reverse_string('LTIMindtree'))
```

# Prime number check

```python
def prime_or_not(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        return False if n%i==0 else True
    
print(prime_or_not(29))
print(prime_or_not(20))
```

# Find duplicate elements in the list

```python
def duplicate_elements(l):
    return list(set(i for i in l if l.count(i)>1))

l = [2,3,4,7,3,4,1,2]
print(duplicate_elements(l))
```

# Handling missing values from df

```python
import pandas as pd

df = pd.DataFrame({"A": [1,2,3,None,6]})
df.fillna(df['A'].mean(), inplace=True)

print(df)
```

# Linear regression from Scratch

```python
import numpy as np

x = np.array([1,2,3,4,5])
y= np.array([2,4,6,8,10])

m,b= 0,0
lr = 0.01

n= len(x)

for _ in range(1000):
    y_pred = m*x + b
    dm = (-2/n) * sum(x*(y - y_pred))
    db = (-2/n) * sum(y - y_pred)
    m -= lr * dms
    b -= lr * db
print("Slope:", m, "Intercept:", b)
```
